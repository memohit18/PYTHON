### Primality Checking

Checking Prime

```python
n = int(input())
d = 2
flag = False
while d < n:
    if n % d == 0:
        flag = True
    d = d+1
    
if flag:
    print("Not prime")
else:
    print("Prime")
    
```

### Nested While Loop

Printing all the prime number between n

```python
n = int(input())
k = 2
while k <= n :
    d = 2
    flag = False
    while d < k:
        if k % d == 0:
            flag = True
        d+=1
    if not flag:
        print(k)
    k+=1
```

### Reverse of a number 

```python

def reverse(n):
    rev = 0
    while n > 0:
        a = n % 10
        rev = rev * 10 + a
        n = n//10
    return rev

n=int(input())
result = reverse(n)
print(result)
```

### Palindrome number 

```python
def checkPalindrome(num):
    temp = num
    rev = 0
    while num > 0:
        digi = num % 10
        rev = rev * 10 + digi
        num = num // 10
    if rev == temp:
        return True
    else:
        return False
    
    
    
num = int(input())
isPalindrome = checkPalindrome(num)
if(isPalindrome):
    print('true')
else:
    print('false')
```

### Nth Fibonacci Number 

```python
n = int(input())
a = 0
b = 1
i = 1
while i <= n:
    s = a+b
    a = b
    b = s
    i+=1
print(a)
```
