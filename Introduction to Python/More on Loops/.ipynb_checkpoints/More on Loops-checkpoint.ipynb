### Check if a Number is Prime

```python
n = int(input())

flag = False
for i in range(2,n,1):
    if n % i == 0:
        flag = True
if flag:
    print('Not Prime')
else:
    print('Prime')
        
```

### Pattern

```python
n = int(input())

for i in range(1,n+1,1):
    for s in range(n-i):
        print(' ',end='')
    for j in range(i,2*i,1):
        print(j,end='')
    for j in range(2*i-2,i-1,-1):
        print(j,end='')
    print()
```

### Binary Pattern

```python
m = int(input())
for i in range(m):
    for j in range(m-i):
        if i % 2 == 0:
            print(1,end='')
        else:
            print(0,end='')
    print()
```

### Number Pyramid

```python
n = int(input())
for i in range(n):
    for s in range(i):
        print(' ',end='')
    for j in range(i+1,n+1):
        print(j,end='')
    print()
    
for i in range(n-2,-1,-1):
    for s in range(i):
        print(' ',end='')
    for j in range(i+1,n+1):
        print(j,end='')
    print()
```

### Diamond of Stars

```python
n = int(input())
n1 = (n+1)//2
n2 = n//2
for i in range(1,n1+1):
    for s in range(1,n1-i+1):
        print(' ',end='')
    for j in range(1,2*i):
        print('*',end='')
    print()
for i in range(n2,0,-1):
    for s in range(1,n1-i+1):
        print(' ',end='')
    for j in range(1,2*i):
        print('*',end='')
    print()
```

### Rectangular Numbers

```python
n = int(input())

for i in range(1,n*2):
    k = n
    if i<= n:
        for j in range(1,n*2):
            print(k,end='')
            if i > j:
                k-=1
            elif i+j >= 2*n:
                k+=1
    
    elif i>= n:
        for j in range(1,n*2):
            print(k,end='')
            if i+j < 2*n:
                k-=1
            elif i <= j:
                k+=1
    print()
```

### Print the Pattern

```python
n = int(input())
for i in range(0,(n+1)//2):
    for j in range(0,n):
        print(2*n*i+j+1,end=' ')
    print()
for i in range((n+1)//2,n):
    for j in range(0,n):
        print(2*n*(n-i)-n+j+1,end=' ')
    print()
```
