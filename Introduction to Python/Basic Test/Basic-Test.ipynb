### Find power of a number 

```python
x = input().split()
power = int(x[0])**int(x[1])
print(power)
```

### Alpha Pattern

```python
n = int(input())
i = 1
while i <= n:
    j = 1
    start_char = chr(ord('A')+i-1)
    while j <= i:
        print(start_char,end='')
        j+=1    
    print()
    i+=1
```

### All prime numbers

```python
n = int(input())
k = 2
while k <= n:
    d = 2
    flag = False
    while d < k:
        if k % d == 0:
            flag = True
        d+=1
    if not flag:
        print(k)
    k+=1
```

### Terms of AP

```python
n = int(input())
count = 1
i = 1
while count <= n:
    ap = (3*i)+ 2
    if ap % 4 != 0:
        print(ap,end=' ')
        count+=1
    i+=1
```

### Repeating numbers

```python
n = int(input())
i = 0
p = 1
while i < n:
    j = 1
    while j<=2**i:
        if p > 9:
            p = 1
        print(p,end='')
        p+=1
        j+=1
    print()
    i+=1
```

### Printing Odd pattern

```python
n = int(input())
i = 1
p = 1
while i < n:
    j = 1
    while j<=i:
        print(p*2-1,end='')
        p+=1
        j+=1
    print()
    i+=1
```
