Inverted Pattern

```python
n = int(input())
i = 1
while i <= n:
    j = 1
    while j <= n-i+1:
        print('*',end = "")
        j+=1
    print()
    i+=1
```

Inverted Number Pattern

```python
n = int(input())
i = 1
while i <= n:
    j = 1
    while j <= n-i+1:
        print(n-i+1,end = "")
        j+=1
    print()
    i+=1
```

Reversed Pattern

```python
n = int(input())
i = 1
while i <= n:
    spaces = 1
    while spaces <= n-i:
        print(' ', end='')
        spaces+=1
    j = 1
    while j <= i:
        print('*',end='')
        j+=1
    print()
    i+=1
```

Mirror Number Pattern

```python
n = int(input())
i = 1
while i <= n:
    spaces = 1
    while spaces <= n-i:
        print(' ', end='')
        spaces+=1
    j = 1
    while j <= i:
        print(j,end='')
        j+=1
    print()
    i+=1
```

Isosceles Pattern

```python
n = int(input())
i = 1
while i <= n:
    spaces = 1
    while spaces <= n-i:
        print(' ',end="")
        spaces+=1
    j = 1
    while j <= i:
        print(j,end='')
        j+=1
    p = i-1
    while p >= 1:
        print(p,end='')
        p-=1
    print()
    i+=1
```

Star Pattern

```python
n = int(input())
i = 1
while i <= n:
    spaces = 1
    while spaces <= n-i:
        print(' ',end='')
        spaces+=1
    j = 1
    while j <= i:
        print('*',end='')
        j+=1
    p = i-1
    while p >= 1:
        print('*',end='')
        p-=1
    print()
    i+=1
```

Triangle of Numbers

```python
n = int(input())
i = 1
while i <= n:
    spaces = 1
    while spaces <= n-i:
        print(' ',end ='')
        spaces+=1
    j = 1
    p = i
    while j <= i:
        print(p,end='')
        j+=1
        p+=1
    p = 2*i-2
    while p >= i:
        print(p,end='')
        p-=1
    print()
    i+=1
```
