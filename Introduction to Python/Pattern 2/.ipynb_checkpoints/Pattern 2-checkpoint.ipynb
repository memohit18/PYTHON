Inverted Pattern

```python
n = int(input())
i = 1
while i <= n:
    j = 1
    while j <= n-i+1:
        print('*',end = "")
        j+=1
    print()
    i+=1
```

Inverted Number Pattern

```python
n = int(input())
i = 1
while i <= n:
    j = 1
    while j <= n-i+1:
        print(n-i+1,end = "")
        j+=1
    print()
    i+=1
```

Reversed Pattern

```python
n = int(input())
i = 1
while i <= n:
    spaces = 1
    while spaces <= n-i:
        print(' ', end='')
        spaces+=1
    j = 1
    while j <= i:
        print('*',end='')
        j+=1
    print()
    i+=1
```

Mirror Number Pattern

```python
n = int(input())
i = 1
while i <= n:
    spaces = 1
    while spaces <= n-i:
        print(' ', end='')
        spaces+=1
    j = 1
    while j <= i:
        print(j,end='')
        j+=1
    print()
    i+=1
```

Isosceles Pattern

```python
n = int(input())
i = 1
while i <= n:
    spaces = 1
    while spaces <= n-i:
        print(' ',end="")
        spaces+=1
    j = 1
    while j <= i:
        print(j,end='')
        j+=1
    p = i-1
    while p >= 1:
        print(p,end='')
        p-=1
    print()
    i+=1
```

Star Pattern

```python
n = int(input())
i = 1
while i <= n:
    spaces = 1
    while spaces <= n-i:
        print(' ',end='')
        spaces+=1
    j = 1
    while j <= i:
        print('*',end='')
        j+=1
    p = i-1
    while p >= 1:
        print('*',end='')
        p-=1
    print()
    i+=1
```

Triangle of Numbers

```python
n = int(input())
i = 1
while i <= n:
    spaces = 1
    while spaces <= n-i:
        print(' ',end ='')
        spaces+=1
    j = 1
    p = i
    while j <= i:
        print(p,end='')
        j+=1
        p+=1
    p = 2*i-2
    while p >= i:
        print(p,end='')
        p-=1
    print()
    i+=1
```

Diamond of stars

```python
n = int(input())
n1 = (n+1)//2
n2 = n//2

i = 1
while i <= n1:
    spaces = 1
    while spaces <= n1-i:
        print(' ',end ='')
        spaces+=1
        
    j = 1
    while j <= 2*i-1:
        print('*',end='')
        j+=1
    print()
    i+=1
    
i = n2
while i >= 1:
    spaces = 1
    while spaces <= n2-i+1:
        print(' ',end ='')
        spaces+=1
        
    j = 1
    while j <= 2*i-1:
        print('*',end='')
        j+=1
    print()
    i-=1
```

Number Pattern

```python
n = int(input())
i = 1
while i <= n:
    j = 1
    while j <= i:
        print(j,end='')
        j+=1
    spaces  = 1
    while spaces <= 2*(n-i):
        print(' ',end='')
        spaces+=1
    j = i
    while j >= 1:
        print(j,end='')
        j-=1
    print()
    i+=1
```

Zeros and Stars Pattern

```python
n = int(input())
i = 1
while i <= n:
    j=1
    while j<= n:
        if j == i:
            print('*',end='')
        else:
            print('0',end='')
        j+=1
    print('*',end='')
    p = j-1
    while p >= 1:
        if p == i:
            print('*',end='')
        else:
            print('0',end='')
        p-=1
    print()
    i+=1
```

Pyramid Number Pattern

```python
n = int(input())
i = 1
while i <= n:
    spaces = 1
    while spaces <= n-i:
        print(' ',end='')
        spaces+=1
    j = 1
    p = i
    while j <= i:
        print(p,end='')
        p-=1
        j+=1
    j = 1
    k = 2
    while j < i:
        print(k,end='')
        j+=1
        k+=1
    print()
    i+=1
```

Arrow pattern

```python
n = int(input())
n1 = (n+1)//2
n2 = n//2

i = 1
while i <= n1:
    spaces = 1
    while spaces <= i-1:
        print(' ',end='')
        spaces+=1
    j = 1
    while j <= i:
        print('* ',end='')
        j+=1
    print()
    i+=1
    
i= 1
while i <= n2:
    spaces = 1
    while spaces <= n2-i:
        print(' ',end='')
        spaces+=1
    j = 1
    while j <= n2-i+1:
        print('* ',end='')
        j+=1
    print()
    i+=1
```
