### Triangle Number Pattern 

```python
n = int(input())
i = 1
while i <= n:
    j = 1
    while j <= i:
        print(i,end = '')
        j+=1
    print()
    i+=1
```

### Reverse Number Pattern

```python
n = int(input())
i = 1
while i <= n:
    j = 1
    p=i
    while j <= i:
        print(p,end = '')
        p-=1
        j+=1
    print()
    i+=1
```

## Character Pattern

```python
# print kth alphabet
k = int(input())
# 'A' + k - 1
x = ord('A')
asciiTarget = x+k-1
targetChar = chr(asciiTarget)
print(targetChar)
```

```python
n = int(input())
i = 1
while i <= n:
    j = 1
    while j <= n:
        #print jth column
        charP= chr(ord('A') + j -1)
        print(charP,end='')
        j+=1    
    print()
    i+=1
```

```python
n = int(input())
i = 1
while i <= n:
    j = 1
    start_char = chr(ord('A') + i -1)
    while j <= n:
        #print jth column
        charP= chr(ord(start_char) + j -1)
        print(charP,end='')
        j+=1    
    print()
    i+=1
```

## Interesting Alphabets 

```python
n = int(input())
i = 1
while i <= n:
    j = 1
    start_char = chr(ord('A') + n - i)
    while j <= i:
        #print jth column
        charP= chr(ord(start_char) + j -1)
        print(charP,end='')
        j+=1    
    print()
    i+=1
```

### Number Pattern 2

```python
n = int(input())
print(1)
i = 1
while i < n:
    j = 0
    while j <= i:
        if j == 0 or j == i:
            print(i,end='')
        else:
            print(0,end='')
        j+=1
    print()
    i+=1
```

### Number Pattern 3 

```python
n = int(input())
print(1)
i = 1
while i < n:
    j = 0
    while j <= i:
        if j == 0 or j == i:
            print(1,end='')
        else:
            print(2,end='')
        j+=1
    print()
    i+=1
```

### Number Pattern
