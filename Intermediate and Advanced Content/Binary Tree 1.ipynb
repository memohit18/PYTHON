```python
class BinaryTreeNode:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
```

```python
def printTree(root):
    if root == None:
        return
    print(root.data)
    printTree(root.left)
    printTree(root.right)
```

```python
def printTreeDetailed(root):
    if root == None:
        return
    print(root.data,end=":")
    if root.left != None:
        print("L",root.left.data,end=",")
    if root.right != None:
        print("R",root.right.data,end="")
    print()
    printTreeDetailed(root.left)
    printTreeDetailed(root.right)
```

```python
def treeInput():
    rootData = int(input())
    if rootData == -1:
        return None
    
    root = BinaryTreeNode(rootData)
    leftTree = treeInput()
    rightTree = treeInput()
    root.left = leftTree
    root.right = rightTree
    return root
```

```python
def numNodes(root):
    if root is None:
        return 0
    
    leftCount = numNodes(root.left)
    rightCount = numNodes(root.right)
    return 1+ leftCount + rightCount
```

```python
root = treeInput()
printTreeDetailed(root)
print(numNodes(root))
```
