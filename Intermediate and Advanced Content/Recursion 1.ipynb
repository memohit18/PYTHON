### Factorial Code


```python
n = int(input())

def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)

fact(n)
```

## Problem solving using Recursion

```python
n = int(input())

def sum_n(n):
    if n == 0:
        return 0
    smallOutput = sum_n(n-1)
    output = smallOutput + n
    return output

sum_n(n)
```

### Power Of A Number 

```python
arr=[int(ele)for ele in input().split()]
x=arr[0]
n=arr[1]

def power(x,n):
    if x==0 and n==0:
        return 1
    elif n == 0:
        return 1
    if x == 0:
        return 0
    else:
        return x*power(x,n-1)
    

print(power(x,n))
```

### Nth Fibonacci Series

```python

def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    fib_n_1 = fib(n-1)
    fib_n_2 = fib(n-2)
    output= fib_n_1+fib_n_2
    return output

n = int(input())
fib(n)
```

```python
def isSorted(a):
    l = len(a)
    if l==0 or l==1:
        return True
    
    if a[0] > a[1]:
        return False
    
    smallerList = a[1:]
    isSmallerLlistSort = isSorted(smallerList)
    
    if isSmallerLlistSort:
        return True
    else:
        return False
    
a=[1,2,3,4,5,6,7,8,9]
print(isSorted(a))
a=[1,2,3,4,9,6,7,8,9]
print(isSorted(a))
```

### Sum of Array

```python
def sumArray(arr):
    if len(arr) == 0:
        return 0
    else:
        return arr[0]+sumArray(arr[1:])

# Main
from sys import setrecursionlimit
setrecursionlimit(11000)
n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
print(sumArray(arr))
```

### Check number in Array

```python
def checkNumber(arr, x):
    if len(arr)==0:
        return 0
    elif arr[0]==x:
        return x
    a=checkNumber(arr[1:],x)
    if a:
        return True
    else:
        return False

# Main
from sys import setrecursionlimit
setrecursionlimit(11000)
n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
x=int(input())
if checkNumber(arr, x):
    print('true')
else:
    print('false')
```

### Is list Sort

```python
def isSortedBetter(a,si):
    l = len(a)
    if si == l-1 or si == l:
        return True
    
    if a[si] > a[si+1]:
        return False
    
    isSmallerPartSorted = isSortedBetter(a,si+1)
    return isSmallerPartSorted
    
a=[1,2,3,4,5,6,7,8,9]
print(isSortedBetter(a,0))
a=[1,2,3,4,50,6,7,8,9]
print(isSortedBetter(a,0))
```

### First Index of Number

```python
def firstIndex(arr, x):
    if len(arr)==0:
        return -1
    if arr[0] == x:
        return 0
    
    smallerList = arr[1:]
    smallerListOutput = firstIndex(smallerList,x)
    
    if smallerListOutput==-1:
        return -1
    else:
        return smallerListOutput+1
    

# Main
from sys import setrecursionlimit
setrecursionlimit(11000)
n=int(input())
arr=list(int(i) for i in input().strip().split(' '))
x=int(input())
print(firstIndex(arr, x))
```

```python
def firstIndex(arr,x,si):
    l = len(arr)
    if si == l:
        return -1
    
    if arr[si] == x:
        return si
    
    smallerListOutput = firstIndex(arr,x,si+1)
    return smallerListOutput
    
    
    
arr=[1,2,3,4,5,6,7,8,9]  
print(firstIndex(arr,7,0))
```

### Last Index Number
