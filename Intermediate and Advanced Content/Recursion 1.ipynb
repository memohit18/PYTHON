### Factorial Code


```python
n = int(input())

def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)

fact(n)
```

## Problem solving using Recursion

```python
n = int(input())

def sum_n(n):
    if n == 0:
        return 0
    smallOutput = sum_n(n-1)
    output = smallOutput + n
    return output

sum_n(n)
```

### Power Of A Number 

```python
arr=[int(ele)for ele in input().split()]
x=arr[0]
n=arr[1]

def power(x,n):
    if x==0 and n==0:
        return 1
    elif n == 0:
        return 1
    if x == 0:
        return 0
    else:
        return x*power(x,n-1)
    

print(power(x,n))
```
